```python
# Import required packages
%matplotlib notebook
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
```

```python
# Edit plot parameters
mpl.rcParams['font.family'] = 'Avenir'
mpl.rcParams['font.size'] = 18

mpl.rcParams['axes.spines.left'] = False
mpl.rcParams['axes.spines.top'] = False
mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['axes.spines.bottom'] = False
```

```python
# Define the Lissajous function
def lissajous(a, b, t):
    x = np.sin(a*t)
    y = np.sin(b*t)
    return x, y
```

```python
# Create figure and add axes
fig = plt.figure(figsize=(3, 3))
ax = fig.add_subplot(111)

# Create an empty plot with variable reference
l_j, = ax.plot([], [], linewidth=2.5)

# Define range shift at each timestep
range_shift = 8*np.pi

# Animation function that updates the data
def animate(i):
    t = np.arange(0 + range_shift*i, range_shift*(1 + i), 0.01)
    x, y = lissajous(2, 1, t)
    l_j.set_data(x, y)

# Use FuncAnimation to animate figure
ani = FuncAnimation(fig=fig, func=animate, frames=range(10), interval=100, repeat=True)

# Set axis limits and remove tick marks
ax.set(xlabel='a = 2\nb = 1', xlim=(-1.05, 1.05), ylim=(-1.05, 1.05), xticks=([]), yticks=([]))

# Make sure animation is fully visible
plt.tight_layout()

# Save and show animation
ani.save('Lissajous_1_1.gif', writer='pillow', fps=10)

# Show animation
plt.show()
```
